```python
X=np.array([[1,2],[5,8],[1.5,1.8],[8,8],[1,0.6],[9,11]])
```

```python
Y=[0,1,0,1,0,1]
```

```python
clf=svm.SVC(kernel='linear',C=1.0)
```

```python
clf.fit(X,Y)
```

```python
w=clf.coef_[0]
```

```python
print(w)
```

```python
a=-w[0]/w[1]
```

```python
xx=np.linspace(0,12)
```

```python
yy=a*xx-clf.intercept_[0]/w[1]
```

```python
h0=plt.plot(xx,yy,'k-',label='non weighted div')
```
